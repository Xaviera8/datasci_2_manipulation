#Data Cleaning With Pandas

## Import Packages


```python
import pandas as pd
import re
import os
```

##Load the Data set


```python
df = pd.read_csv('healthcare_data_cleaning.csv')
df
```

##Check for any missing values in the dataset

```python
print(df.isnull().sum())
```

## Remove any missing values from the dataset

```python
df.dropna(inplace=True)
print(df.isnull().sum())
```

## Remove any duplicate rows and columns, if they exist

```python
df.drop_duplicates(inplace=True)
df
```

## Clean Colums Names????

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(' ', '_')
df
```

#Data Transformation


## Create new columns based on existing ones (e.g., if you have a 'birth_date' column, create an 'age' column).

```python
df.columns
```

## Aggregate data using groupby and compute summary statistics.

```python
df['Weight_(in_kg)'] = pd.to_numeric(df['Weight_(in_kg)'], errors='coerce')
df['Weight_(in_kg)'].fillna(df['Weight_(in_kg)'].median(), inplace=True)


grouped_data = df.groupby('State_of_Residence').agg({
    'Weight_(in_kg)': ['mean']  # Average Age
})
print(grouped_data)
```

## Use Pivot Tables or Cross Tabulations for Multidimensional Analysis

```python
df['Patient_Age'] = pd.to_numeric(df['Patient_Age'], errors='coerce')
df = df.dropna(subset=['Patient_Age'])

df['Weight Group in KG'] = pd.cut(
    df['Weight_(in_kg)'],
    bins = (66, 68, 70, 72),
    labels = ['66 - 68', '68 - 70', '70 - 72'])
pivottable = df.pivot_table(index = 'Weight Group in KG', columns = 'Gender', values ='Patient_Age', aggfunc = 'mean' )

print (pivottable)


```
